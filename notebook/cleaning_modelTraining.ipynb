```python
x_train['feedback']=x_train['feedback'].apply(lambda x:re.sub('[^a-z A-z 0-9-]+', '',x)) 
## Remove the stopswords
x_train['feedback']=x_train['feedback'].apply(lambda x:" ".join([y for y in x.split() if y not in stopwords.words('english')]))
## Remove url 
x_train['feedback']=x_train['feedback'].apply(lambda x: re.sub(r'(http|https|ftp|ssh)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', '' , str(x)))
## Remove any additional spaces
x_train['feedback']=x_train['feedback'].apply(lambda x: " ".join(x.split()))
```

```python

import gensim.downloader as api
wv = api.load('word2vec-google-news-300')

```

```python
# Function to convert text to average word vector
def text_to_vector(text, model):
    words = text.split()
    word_vectors = []
    
    for word in words:
        if word in model:
            word_vectors.append(model[word])
    
    if len(word_vectors) == 0:
        return np.zeros(300)  # Return zero vector if no words found
    
    return np.mean(word_vectors, axis=0)

```

```python
# Apply to train and test sets
x_train_feedback_vectors = x_train['feedback'].apply(lambda x: text_to_vector(x, wv))
x_test_feedback_vectors = x_test['feedback'].apply(lambda x: text_to_vector(x, wv))

```

```python
# Convert to DataFrame with column names
feedback_train_df = pd.DataFrame(x_train_feedback_vectors.tolist(), 
                                  columns=[f'feedback_vec_{i}' for i in range(300)],
                                  index=x_train.index)
feedback_test_df = pd.DataFrame(x_test_feedback_vectors.tolist(), 
                                 columns=[f'feedback_vec_{i}' for i in range(300)],
                                 index=x_test.index)

```

```python
# Drop original feedback column and concatenate vectors
x_train = x_train.drop('feedback', axis=1)
x_test = x_test.drop('feedback', axis=1)

x_train = pd.concat([x_train, feedback_train_df], axis=1)
x_test = pd.concat([x_test, feedback_test_df], axis=1)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
stsc=ColumnTransformer([
    ('std_scaler', StandardScaler(), [col for col in x_train.columns ])
], remainder='passthrough', verbose_feature_names_out=False)
x_train=stsc.fit_transform(x_train)
x_test=stsc.transform(x_test)
columns=stsc.get_feature_names_out()
x_train=pd.DataFrame(x_train,columns=columns)
x_test=pd.DataFrame(x_test,columns=columns)
```

```python
x_train
```

```python
# 1. Fix invalid scores (-1) by clipping them to the 1-5 range
y_train = y_train.clip(1, 5)
y_test = y_test.clip(1, 5)

# 2. Shift labels from 1-5 to 0-4 for the Softmax layer
y_train_final = y_train - 1
y_test_final = y_test - 1
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization

model = Sequential([
    # Input Layer: 331 features
    Dense(512, activation='relu', input_shape=(x_train.shape[1],)),
    BatchNormalization(),
    Dropout(0.3),
    
    # Hidden Layers
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.2),


    Dense(512, activation='relu'),
    BatchNormalization(),
    Dropout(0.2),
    
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.2),
    
    Dense(128, activation='relu'),
    Dropout(0.2),
    
    # Output Layer: 5 neurons for labels 0-4
    Dense(5, activation='softmax')
])



```

```python
model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=8,
    restore_best_weights=True
)
```

```python
# Execute training
history = model.fit(x_train, y_train_final, 
                    epochs=60, 
                    batch_size=64, 
                    validation_data=(x_test, y_test_final),callbacks=[early_stop])
```

```python
from sklearn.metrics import classification_report, f1_score, confusion_matrix
import numpy as np

# 1. Get the raw probability predictions
y_pred_probs = model.predict(x_test)

# 2. Convert probabilities to class labels (0-4)
y_pred = np.argmax(y_pred_probs, axis=1)

# 3. Calculate the official HackerEarth Score (Macro F1)
# Note: y_test_final is your shifted 0-4 ground truth
score = 100 * f1_score(y_test_final, y_pred, average="macro")
print(f"HackerEarth Evaluation Score: {score:.2f}")

# 4. Detailed breakdown (Precision, Recall, F1 for each class)
print("\nClassification Report:")
print(classification_report(y_test_final, y_pred, target_names=['Score 1', 'Score 2', 'Score 3', 'Score 4', 'Score 5']))
```

```python
from sklearn.metrics import recall_score, precision_score
recall = recall_score(y_test_final, y_pred, average='macro')
precision = precision_score(y_test_final, y_pred, average='macro')
print(f"Recall: {recall:.4f}, Precision: {precision:.4f}")
```
